```python
import cv2 as cv 
from google.colab.patches import cv2_imshow 
from skimage import io 
import matplotlib.pyplot as plt 
import numpy as np
 
# Load gambar dari Google Drive
img = cv.imread('/content/drive/MyDrive/PCVK/Image/Cat01.jpeg')

# Tampilkan gambar
plt.imshow(img) 
```

```python
# 2. OpenCV membaca image dan menyimpan dalam channel warna BGR (Blue Green Red)
#    Tampilkan nilai pixel dan image

# cvtColor merupakan function warna yang disediakan pada cv
img2 = cv.cvtColor(img,cv.COLOR_BGR2RGB) # koversi channel BGR -> RGB
plt.imshow(img2)
```

```python
# 3. Menampilkan citra Grayscale, melakukan resizing, melakukan Flipping, 
#    Menyimpan citra hasil

# Ubah gambar dalam bentuk grayscale
img_gray = cv.imread('/content/drive/MyDrive/PCVK/Image/Cat01.jpeg', cv.IMREAD_GRAYSCALE)

# Tampilkan gambar
plt.imshow(img_gray)
```

```python
# Parameter cmap digunakan untuk merepresentasikan colormap
plt.imshow(img_gray, cmap='gray')
```

```python
plt.imshow(img_gray, cmap='magma')
```

```python
# function cv.resize digunakan untuk merubah resolusi dari gambar
# Resolusi dari gambar berikut dirubah menjadi ukuran 512 x 1024
img4 = cv.resize(cv.cvtColor(img, cv.COLOR_BGR2RGB), (512, 1024))
plt.imshow(img4)
```

```python
# function cv.flip digunakan untuk mebalikkan gambar
img5 = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),0)

# Simpan file image
#cv.imwrite('/content/drive/MyDrive/PCVK/Images',img5)

#Tampilkan plot dengan ukuran canvas yang lebih besar
fig = plt.figure(figsize=(10,10)) #Ubah-ubah ukuran (10,10) sesuai kebutuhan
ax = fig.add_subplot(111)
ax.imshow(img5)
```

```python
# 1. Tampilkan image dalam channel Red-Green dan Green Blue saja!

# Load gambar
img_kitten = cv.imread('/content/drive/MyDrive/PCVK/Image/Cat01.jpeg')
# Konversi gambar dari BGR ke RGB
img_kitten2 = cv.cvtColor(img_kitten, cv.COLOR_BGR2RGB)
# Buat variable RGB untuk dilakukan split warna pada gambar
R,G,B = cv.split(img_kitten2)
# Method zeros digunakan untuk membuat array kosong
zeros = np.zeros(img_kitten2.shape[:2], dtype="uint8")
# Tampilkan warna merah hijau
red_green = cv.merge([R,G,zeros])
plt.imshow(red_green)
```

```python
green_blue = cv.merge([zeros,G,B])
plt.imshow(green_blue)
```

```python
# 2. Tampilkan image baris ke 10-100, kolom 10-100!

img_kitten3 = cv.imread('/content/drive/MyDrive/PCVK/Image/Cat01.jpeg')
img_kitten4 = cv.cvtColor(img_kitten3, cv.COLOR_BGR2RGB)
crop = img_kitten4[10:200, 10:200]
plt.imshow(crop)
```

```python
# 3. Tampilkan image baris ke 5-30 semua kolom, channel Green saja

green = cv.merge([zeros,G,zeros])
y=5
x=0
h=30
w=green.shape[1]
crop = green[y:h, x:w]
plt.imshow(crop)
```

```python
# 4. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!

from random import randrange
img_kitten5 = cv.imread('/content/drive/MyDrive/PCVK/Image/Cat01.jpeg')
img_kitten6 = cv.cvtColor(img_kitten5, cv.COLOR_BGR2RGB)
tinggi = img_kitten6.shape[0]                            
lebar = img_kitten6.shape[1]  
for x in range(5):
  cv.rectangle(img_kitten6, (randrange(lebar),randrange(tinggi)), (randrange(lebar),randrange(tinggi)), (randrange(255),randrange(255),randrange(255)), 5)
plt.imshow(img_kitten6)
```

```python
# 5. Tampilkan image dengan posisi terbalik!

img_kitten7 = cv.imread('/content/drive/MyDrive/PCVK/Image/Cat01.jpeg')
img_kitten8 = cv.flip(cv.cvtColor(img_kitten7, cv.COLOR_BGR2RGB), 0)
plt.imshow(img_kitten8)
```

```python
# 6. Buat rectangle dan circle pada bagian wajah dari image kitten!

img_kitten9 = cv.imread('/content/drive/MyDrive/PCVK/Image/Cat01.jpeg')
img_kitten10 = cv.cvtColor(img_kitten9, cv.COLOR_BGR2RGB)
cv.circle(img_kitten10, (240,118), 50, (255,0,0), 5)
cv.rectangle(img_kitten10, (310,170), (150,50), (0,255,0), 5)
plt.imshow(img_kitten10)
```
